1. Ajuste um modelo para prever log(renda) considerando todas as covariáveis disponíveis.
    - Utilizando os recursos do Patsy, coloque as variáveis qualitativas como *dummies*.
    - Mantenha sempre a categoria mais frequente como casela de referência
    - Avalie os parâmetros e veja se parecem fazer sentido prático.

```python
# Fórmula para o modelo
formula = (
    'np.log(renda) ~ idade + C(sexo, Treatment(1)) + C(posse_de_veiculo, Treatment(1)) + '
    'C(posse_de_imovel, Treatment(1)) + qtd_filhos + C(tipo_renda, Treatment(2)) + '
    'C(educacao, Treatment(2)) + C(estado_civil, Treatment(2)) + '
    'C(tipo_residencia, Treatment(2)) + qt_pessoas_residencia + tempo_emprego'
)

# Gerar as matrizes de design para y e X
y,X = dmatrices(formula, data=df, return_type="dataframe")

# Exibir as primeiras linhas das matrizes
print("Variáveis independentes (X):")
print(X.head())
print("\nVariável dependente (y):")
print(y.head())

reg1 = sm.OLS(y,X).fit()
reg1.summary()
```

2. Remova a variável menos significante e analise:
    - Observe os indicadores que vimos, e avalie se o modelo melhorou ou piorou na sua opinião.
    - Observe os parâmetros e veja se algum se alterou muito.

```python
formula_2 = (
    'np.log(renda) ~ idade + C(sexo, Treatment(1)) + C(posse_de_veiculo, Treatment(1)) + '
    'C(posse_de_imovel, Treatment(1)) + qtd_filhos + C(tipo_renda, Treatment(2)) + '
    'C(estado_civil, Treatment(2)) + '
    'qt_pessoas_residencia + tempo_emprego'
)

y,X = dmatrices(formula_2, data=df, return_type='dataframe')

print("Variáveis independentes (X):")
print(X.head())
print("\nVariável dependente (y):")
print(y.head())

reg2 = sm.OLS(y,X).fit()
reg2.summary()
```

No segundo modelo o R² subiu um pouco e a maioria das variaveis estão menores que 5%.
