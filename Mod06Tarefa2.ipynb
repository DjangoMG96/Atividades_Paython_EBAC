```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np


%matplotlib inline

csv = 'sinasc_RO_2019.csv'

df = pd.read_csv(csv)
df['PESO'] = df['PESO'] / 1000
df[['IDADEMAE', 'PESO']].describe()
```

####  1) Na base do SINASC do DataSUS, faça um boxplot da idade da mãe.

```python
df[['IDADEMAE']].boxplot()
plt.show()
```

####  2) Faça um boxplot do peso do bebê

```python
df[['PESO']].boxplot()
plt.show()
```

####  3) Faça um histograma da idade da mãe

```python
# Regra de Sturges para identificar o melhor numero de bins
len(df['IDADEMAE']) # 27028
k = 1 + np.log2(27028)
K = int(k) # 15

df['IDADEMAE'].plot.hist(bins = K)
plt.show()
```

#### 4) Faça um histograma do peso do bebê

```python
sns.displot(data = df['PESO']
           bins = K)
```
